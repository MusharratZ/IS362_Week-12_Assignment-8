```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data.csv')
```

```python
# Subset the DataFrame
subset_df = df[['poisonous', 'odor', 'cap-color']]  # 'cap-color' is an example, choose any other column you prefer
```

```python
# Rename columns
subset_df.columns = ['Edible', 'Odor', 'Cap_Color']
```

```python
# Replace codes with numeric values
subset_df.loc[:, 'Edible'] = subset_df['Edible'].replace({'e': 0, 'p': 1})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of data for each column
sns.histplot(subset_df['Edible'], bins=2, kde=False)
plt.title('Distribution of Edible and Poisonous Mushrooms')
plt.xlabel('Edible (0) / Poisonous (1)')
plt.show()

sns.countplot(x='Odor', hue='Edible', data=subset_df)
plt.title('Edible vs. Poisonous Mushrooms by Odor')
plt.show()

sns.countplot(x='Cap_Color', hue='Edible', data=subset_df)
plt.title('Edible vs. Poisonous Mushrooms by Cap Color')
plt.show()
```
